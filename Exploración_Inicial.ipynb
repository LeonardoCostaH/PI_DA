```python
import pandas as pd
import numpy as np
```

Lectura y conversión del archivo en dataframe

```python
df_accidentes = pd.read_csv("datasets/AccidentesAviones.csv")
df_accidentes.head()
```

en las celdas vacías hay una señal de *"?"*, les voy a cambiar por un *nan*, para facilitar el análisis

```python
df_accidentes = df_accidentes.replace("?", np.nan)
df_accidentes.info()
```

verificar si hay valores ausentes en las columnas

```python
df_accidentes.isnull().sum()
```

verificar si hay líneas duplicadas

```python
df_accidentes = df_accidentes.drop_duplicates()
df_accidentes.info()
```

No hay valores duplicados

```python
df_accidentes.isnull().sum()
```

Cambio de formato de hora

```python
import pandas as pd
from datetime import datetime


def formatar_hora(hora):
    if pd.notna(hora) and isinstance(hora, str):
        try:
            hora_formatada = datetime.strptime(hora, '%H%M').strftime('%H:%M')
            return hora_formatada
        except ValueError:
            return hora  # Mantenha o valor original se não for possível converter
    else:
        return hora  # Mantenha o valor original se não for uma string ou é nulo (NaN)


# Aplicar a função à coluna "Hora Declarada"
df_accidentes['HORA declarada'] = df_accidentes['HORA declarada'].apply(formatar_hora)

# Resultado
df_accidentes
```

cambio de formato de fecha

```python
# Convierte la columna "fecha" al formato de fecha de pandas
df_accidentes['fecha'] = pd.to_datetime(df_accidentes['fecha'], format='%B %d, %Y')


df_accidentes
```

```python
df_accidentes = df_accidentes.drop("Unnamed: 0", axis=1)
df_accidentes
```

separar la columna de fecha en dia, mes y año

```python
def separar_fecha(df):
    # Convierte la columna "fecha" en formato de cadena (string) si no lo está
    df['fecha'] = df['fecha'].astype(str)
    
    # Divide la columna "fecha" en tres columnas: "año", "mes" y "dia"
    df[['año', 'mes', 'dia']] = df['fecha'].str.split('-', expand=True)
    
    # Convierte las columnas "año", "mes" y "dia" en tipo numérico (entero)
    df['año'] = df['año'].astype(int)
    df['mes'] = df['mes'].astype(int)
    df['dia'] = df['dia'].astype(int)

# Llama a la función para separar la fecha en el DataFrame
separar_fecha(df_accidentes)

# Muestra el DataFrame resultante con las nuevas columnas
df_accidentes
```

renombrando las columnas para mejor comprensión

```python
df_accidentes = pd.DataFrame({
    "Fecha": df_accidentes["fecha"],
    'Año': df_accidentes['año'],
    'Mes': df_accidentes['mes'],
    'Dia': df_accidentes['dia'],
    'Hora Declarada': df_accidentes['HORA declarada'],
    'Ruta': df_accidentes['Ruta'],
    'Ruta(información extra)': df_accidentes['route'],
    'Operadora': df_accidentes['OperadOR'],
    'Número del Vuelo': df_accidentes['flight_no'],
    'Tipo de Aeronave': df_accidentes['ac_type'],
    'Registro': df_accidentes['registration'],
    'Número de Série': df_accidentes['cn_ln'],
    'Total a Bordo': df_accidentes['all_aboard'],
    'Pasajeros a Bordo': df_accidentes['PASAJEROS A BORDO'],
    'Tripulación a Bordo': df_accidentes['crew_aboard'],
    'Total de Fatalidades': df_accidentes['cantidad de fallecidos'],
    'Fatalidades de Pasajeros': df_accidentes['passenger_fatalities'],
    'Fatalidades de la Tripulación': df_accidentes['crew_fatalities'],
    'Fatalidades en Solo': df_accidentes['ground'],
    'Resumen': df_accidentes['summary']
})
```
